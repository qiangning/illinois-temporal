```python
from matplotlib import pyplot as plt
from scipy.signal import savgol_filter

import numpy as np
filepath = "/home/qning2/Research/illinois-temporal/logs/tmp/tempRel/tbdense/CoDL_PinBase/"
fname_prefix = "eeTempRelClsNoVagSam"
fname_middle = "win3_1mdl_global_resp_soft"
#fname_middle = "win3_2mdl_global_resp_soft_lambda0.1"
AllF1_samMod_rel = []
AllF1_samMod_doc = []
AllF1_samMod_imp = []
AllF1_samMod_rel_std = []
AllF1_samMod_doc_std = []
AllF1_samMod_imp_std = []
def readF1fromFile(fname):
    print(fname)
    with open(fname) as f:
        line = f.readline()
        while line and "    EVALUATING EVENT TEMPREL CLASSIFICATION (MODE=1)    " not in line:
            line = f.readline()
        while line and "F-score = " not in line:
            line = f.readline()
        line = line.split(" ");
        return float(line[2])

    

for sr in range(1,11):
# for sr in range(1,10,2):
    f1_samMod_rel = []
    f1_samMod_doc = []
    for seed in range(1,51):
    #for seed in [1, 2, 5, 6, 9, 10, 13, 14]:
    #for seed in [3, 4, 7, 8, 11, 12, 15, 16]:
        fname_samMod_rel = filepath+"seed"+str(seed)+"/"+fname_prefix+"_sr"+("%.1f" %(1.0*sr/10))+"_"+fname_middle+"_sm0.txt"
        fname_samMod_doc = filepath+"seed"+str(seed)+"/"+fname_prefix+"_sr"+("%.1f" %(1.0*sr/10))+"_"+fname_middle+"_sm3.txt"
        if(seed!=0):
            fname_samMod_rel = fname_samMod_rel.replace(".txt","_sd"+str(seed)+"_max1.txt")
            fname_samMod_doc = fname_samMod_doc.replace(".txt","_sd"+str(seed)+"_max1.txt")
        f1_samMod_rel.append(readF1fromFile(fname_samMod_rel))
        f1_samMod_doc.append(readF1fromFile(fname_samMod_doc))
    imp = np.subtract(f1_samMod_rel,f1_samMod_doc)
    AllF1_samMod_rel.append(np.mean(f1_samMod_rel))
    AllF1_samMod_doc.append(np.mean(f1_samMod_doc))
    AllF1_samMod_imp.append(np.mean(imp))
    AllF1_samMod_rel_std.append(np.std(f1_samMod_rel))
    AllF1_samMod_doc_std.append(np.std(f1_samMod_doc))
    AllF1_samMod_imp_std.append(np.std(imp))
print [x*100 for x in np.subtract(AllF1_samMod_rel,AllF1_samMod_doc)]
print [x*100 for x in AllF1_samMod_rel_std]
print [x*100 for x in AllF1_samMod_doc_std]
```
