```python
import numpy as np
import copy
import math
import matplotlib.pyplot as plt
import pickle as pkl
import random
from scipy.signal import savgol_filter

```

```python
allIk = pkl.load(open("Chunker_d10_seq100_rep1000.pkl"))
NumExp = 1000
NumSeq = 100
seqlength = 10
b2 = np.empty([allIk.shape[0],allIk.shape[1]-1])
for i in range(0,allIk.shape[0]):
    for j in range(1,allIk.shape[1]):
        b2[i,j-1] = allIk[i,j-1]-allIk[i,j]
```

```python
num_topo_all = pkl.load(open("N10-rep300.pkl"))
num_nodes = 10
num_edges = range(5,35)
b = np.empty([num_topo_all.shape[0],num_topo_all.shape[1]-1])
for i in range(0,num_topo_all.shape[0]):
    for j in range(1,num_topo_all.shape[1]):
        b[i,j-1] = math.log(1.0*num_topo_all[i,j-1]/num_topo_all[i,j],2)
```

```python
plt.figure(figsize=(6,5))
maxnum = 10
srl_offset = -1.5
chunker_offset = -0.72
default_offset = 0.75
x = range(1,maxnum+1)
y = [-math.log(math.factorial(maxnum-t)) for t in x]
y = np.diff(y)
plt.plot([100.0*t/len(x) for t in x[1:]],[t+srl_offset for t in y],'k') # srl
plt.plot([2.0*e/num_nodes/(num_nodes-1)*100 for e in num_edges[1:]],savgol_filter(np.mean(b,axis=0),7,1),'k--') # DAG
plt.plot([100.0*x/seqlength for x in range(1,seqlength)],[t+chunker_offset for t in savgol_filter(np.mean(b2,axis=0),7,1)],'k-.') # chunker
plt.plot(range(100), [default_offset]*100,'k:')
plt.legend(["Semantic Roles","DAG","BIO Chunking","No Structure"])
plt.xlabel("Completeness ($100k/d$ %)")
plt.ylabel("bit")
plt.title("$I_{k}-I_{k-1}$")
plt.tight_layout()

plt.xlabel("Completeness ($100k/d$ %)")
plt.ylabel("bit")
plt.title("$I_{k}-I_{k-1}$ (shifted)")
plt.rcParams.update({'font.size':16})
plt.ylim(ymin=0,ymax=0.8)
plt.xlim((20,70))
plt.rcParams.update({'font.size':16})
plt.tight_layout()
plt.gca().axes.get_yaxis().set_visible(False)

plt.savefig("AllCompare.pdf")
plt.show()
```
